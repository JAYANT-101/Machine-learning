```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC 
from sklearn.model_selection import train_test_split

StandardScaler=StandardScaler()
X_train, X_test, y_train, y_test=train_test_split(data, y, test_size=0.3, random_state=42)
```

```python
X_train=StandardScaler.fit_transform(X_train)
X_test=StandardScaler.fit_transform(X_test)
```

```python
lin_svc=LinearSVC(loss="hinge", C=5, dual=True, random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lin_svc, X_train, y_train).mean()
```

With Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf =RandomForestClassifier(n_estimators=100, random_state=42)
forest_clf.fit(X_train, y_train)
forest_clf.predict(X_train)
```

```python
from sklearn.model_selection import cross_val_score
a= cross_val_score(forest_clf, X_train, y_train, cv=10, scoring="accuracy").mean()
a
```
