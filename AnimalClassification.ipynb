Transforming the .jpeg in to vectors 

```python
import os 
path=r'datasets/Animal'
l=[]
for root,dir,files in os.walk(path):
    for file in files:
        first_word=file.split('-',1)
        
        l.append(first_word[0])
        print(first_word[0])
```

```python
uniqe_values=list(set(l))
print(uniqe_values)
```

```python
num_lists=[]
for i in range(len(uniqe_values)):
    num_lists.append(i)
```

```python
identety_dic={}
for uniqe_value,num_list in zip(uniqe_values,num_lists):
    identety_dic[uniqe_value]=num_list
```

```python
identety_dic
```

```python
from matplotlib.image import imread
import cv2
import numpy as np
path=r'datasets/Animal'
path2=r'datasets/Animal/'
data_list=[]
target_list=[]
for root,dir,files in os.walk(path):
    for file in files:
        first_word=file.split('-',1)
        Image=f"{path2}{file}"
        convert_image =cv2.imread(Image,cv2.IMREAD_GRAYSCALE)
        data_list.append(cv2.resize(convert_image, (28, 28)))
        target_list.append(identety_dic[first_word[0]])
```

```python
dataset=np.array(data_list)
final_target=np.array(target_list)
print(dataset.shape)
print(final_target.shape)
```

```python
final_data=np.reshape(dataset, (dataset.shape[0], -1))
```

```python
final_data.shape
```

Done now lets make test and traning sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_data, final_target, test_size=0.2, random_state=42)
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf=RandomForestClassifier(random_state=42)
```

```python
from sklearn.model_selection import cross_val_predict
y_probas_forest=cross_val_predict(forest_clf,X_train,y_train,cv=3)
```

```python
y_probas_forest
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_train,y_probas_forest)
cm
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.astype("float64"))
cross_val_score(forest_clf, X_train_scaled, y_train, cv=3, scoring="accuracy")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
y_train_pred=cross_val_predict(forest_clf,X_train_scaled,y_train,cv=3)
ConfusionMatrixDisplay.from_predictions(y_train,y_train_pred)
plt.show
```

```python
ConfusionMatrixDisplay.from_predictions(y_train,y_train_pred,normalize="true",values_format=".0%")
```

```python
sample_weight = (y_train_pred != y_train)
plt.rc('font', size=10)  # extra code
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, sample_weight=sample_weight,normalize="true", values_format=".0%")
plt.show()
```
